```python
from lattice import Lattice2D, BrickwallLatticeGeometry, SimulationParameters, RectangularLatticeGeometry
import numpy as np

omega = 3 * 2*np.pi / 3.5
params = SimulationParameters(1, lambda t:np.sin(omega*t), np.array([0, -1]), 1e-3, 10, 1, 0.5, 1)
l = Lattice2D(BrickwallLatticeGeometry((7, 14)), params)
```

```python
l.evolve(solver="rk4", force_reevolve=True)
```

```python
l.save_lattice_animation("test_grad_bw.gif", sample_every=100, fps=7)
```
